Tensor'den numpy array'e dönüşümler şu şekildedir.

*   **torch.from_numpy():** numpy'dan tensor'e dönüşüm
*   **numpy():** tensor'den numpy'a dönüşüm



```python
#random numpy array
array = np.random.rand(2,2)
print("{} {}\n".format(type(array),array))

#from numpy to tensor
from_numpy_to_tensor = torch.from_numpy(array)
print("{}\n".format(from_numpy_to_tensor))

# from tensor to numpy
tensor = from_numpy_to_tensor
from_tensor_to_numpy = tensor.numpy()
print("{} {}\n".format(type(from_tensor_to_numpy),from_tensor_to_numpy))
```

**Basic Math with Pytorch**

Pytorch'ta input dimension belirlenmek zorunda değildir. Dolayısıyla input dimension is not valid şeklinde bir uyarıyla karşılaşılmayacaktır.

Ancak herhangi bir tensor farklı bir size'a getirilmek istendiğinde **resize()** methodu kullanılır.

(a ve b bir tensor)

Toplama **torch.add(a,b)** methodu ile yapılır. (a+b)

Çıkarma **a.sub(b)** methodu ile yapılır. (a-b)

Eleman düzeyinde çarpım **torch.mul(a,b)** methodu ile yapılır.

Eleman düzeyinde bölüm **torch.div(a,b) = a / b** methodu ile yapılır.

Ortalama -> **a.mean()**

Standart Deviation ->  **(std): a.std()**

```python
#create tensor 
tensor = torch.ones(3,3)
print("\n",tensor)
```

```python
#Resize
print("{}{}\n".format(tensor.view(9).shape,tensor.view(9)))
```

```python
#Addition
print("Addition: {}\n".format(torch.add(tensor,tensor)))
```

```python
#Subtraction
print("Subtraction: {}\n".format(tensor.sub(tensor)))
```

```python
#Element wise multiplication
tensor = torch.Tensor([1,2,3,4,5])
print("Element wise multiplication: {}\n".format(torch.mul(tensor,tensor)))
```

```python
#Element wise division
print("Element wise division: {}\n".format(torch.div(tensor,tensor)))
```

```python
#Mean
tensor = torch.Tensor([15,25,35,40,50])
print("Mean: {}".format(tensor.mean()))
```

```python
#Standart deviation (std)
print("std: {}".format(tensor.std()))
```

**Variables**

Variable'lar gradient'leri içerisinde toplayan, biriktiren, depolayan yapılardır.

Neural network'teki backward propagation yapısında parametreler güncellenirken gradient'ler hesaplanır.

Gradientler'i handle edebilmek için variable yapısı kullanılır.

Backward propagation yapabilmek için pytorch'ta variable yapısına ihtiyaç vardır.

Basic backward propagation işlemi;

y = x^2 denklemi,

x = [2,4]'ten oluşsun.

Denklem hesaplandığında,

 y = [4,16] 

Loss değerlerinin toplanıp coss değerinin elde edilmesi -> o = (1/2)sum(y) = (1/2)sum(x^2)

o'nun x'e göre türevinin alınması için;

o'nun y'ye göre türevi alınıp, y'nin de x'e göre türevi alınır

∂o/∂x = (∂o/∂y)*(∂y/∂x) = x = [2,4]


```python
#import variable from pytorch library
from torch.autograd import Variable

#define variable
var = Variable(torch.ones(3), requires_grad = True)
var
```

Tensor kütüphanesi çağrılıp 1'lerden oluşan bir tensor yaratılır.

```python
array = [2,4]
tensor = torch.Tensor(array)
x = Variable(tensor, requires_grad = True)
y = x**2
print(" y =  ",y)
```

y = x^2 denklemi hesaplanır.

```python
o = (1/2)*sum(y)
print(" o =  ",o)
```

o değeri hesaplanır.

```python
o.backward() 
```

x'e kadar backward yapılır. Çünkü o, y'den türemekte, y de x'den türemektedir.

```python
print("gradients: ",x.grad) #∂o/∂x = x
```
